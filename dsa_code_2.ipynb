```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("all_stocks_5yr.csv")
```

```python
df = df.dropna()
```

```python
df.duplicated().sum()
```

```python
df["date"] = pd.to_datetime(df["date"])
```

```python
df["Name"].unique()
```

```python
df["Name"].nunique()
```

```python
apple = df[df["Name"] == "AAPL"]
```

```python
apple.head()
```

```python
apple.shape
```

```python
apple.describe()
```

```python
plt.figure(figsize=(12,5))
plt.plot(apple["date"], apple["close"])
plt.title("Apple Closing Price")
plt.xlabel("Date")
plt.ylabel("Closing Price")
plt.show()
```

```python
plt.figure(figsize=(12,5))
plt.plot(apple["date"], apple["volume"])
plt.title("Apple Trading Volume")
plt.xlabel("Date")
plt.ylabel("Volume")
plt.show()
```

```python
plt.figure(figsize=(12,5))
plt.plot(apple["date"], apple["high"], label="High")
plt.plot(apple["date"], apple["low"], label="Low")
plt.title("Apple High vs Low Price")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12,5))
plt.plot(apple["date"], apple["open"], label="Open")
plt.plot(apple["date"], apple["close"], label="Close")
plt.title("Opening vs Closing Price")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12,5))
plt.plot(apple["date"], apple["high"], label="High")
plt.plot(apple["date"], apple["low"], label="Low")
plt.title("High vs Low Price")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.show()
```

```python
apple["Moving_Average"] = apple["close"].rolling(30).mean()
```
